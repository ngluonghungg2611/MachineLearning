```python
from datetime import datetime
import cv2
import numpy as np

#Load pre-trained model
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cam = cv2.VideoCapture(0)

while True:
    ret, image = cam.read()
    image2 = image.copy()
    
    if not ret:
        print('Failed to open webcam')
        break
        
    #Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    #Detect face
    face = face_cascade.detectMultiScale(gray, 1.1, 10)
    
    #Các tham số bao gồm hình ảnh, tọa độ đỉnh của HCN,
    #tọa độ thứ 2 nằm đối diện với đỉnh 1, 
    #Màu (B, G, R), và độ dày của hình chữ nhật
    for (x, y, w, h) in face:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    cv2.imshow('Test', image)
    
    key = cv2.waitKey(1) & 0xFF
    
    #if space key is pressed
    if(key == 32):
        try:
            #Crop frontal face
            frontal_face = image2[y:y+h, x:x + w]
            
            #Resize to a small resolution without changing ratio
            frontal_face = cv2.resize(frontal_face, (120, 120), interpolation=cv2.INTER_LINEAR_EXACT)
            
            #Set file name
            now = datetime.now()
            current_time = now.strftime("%d.%M.%y-%H.%M.%S")
            img_name = "img_{}.jpg".format(current_time)
            
            #save file and crop to 80x120px
            cv2.imwrite(img_name, frontal_face[:,20:100])
            print('Taken picture: {}'.format(img_name))
        except:
            print("Face not found")
    #if escape is pressed
    if(key == 27):
        break

cam.release()
cv2.destroyAllWindows()
```

```python
from datetime import datetime
import cv2
import numpy as np

#Load pre-trained model
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

#Read image in color mode
img = cv2.imread('test2.jpg', -1)

img2 = img.copy()

#Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
#Detect faces
face = face_cascade.detectMultiScale(gray, 1.1, 3)

i = 0

for (x, y, w, h) in face:
    #Draw rectangle around faces
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    try:
        #Take the faces
        frontal_face = img2[y:y+h, x:x + w]

        #Resize faces
        frontal_face = cv2.resize(frontal_face, (120, 120), interpolation=cv2.INTER_LINEAR_EXACT)

        #Set name of image file and save

        img_name = "img_{}.jpg".format(i)
        cv2.imwrite(img_name, frontal_face[:,20:100])
        print('Taken picture: {}'.format(img_name))
        i += 1
    except:
        print("Face not found")
    
cv2.imshow('Test', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
