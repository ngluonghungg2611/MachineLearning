```python
import cv2
import numpy as np 
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('Bienso.jpg')
# Chuyn nh sang gray image
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Chuyn sang nh nh phn
_, imgBi = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
# Chuyn nh sang gray image
imgGrayBi = cv2.cvtColor(imgBi, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(16, 4))
plt.subplot(141), plt.imshow(img), plt.title('Origin Image')
plt.subplot(142), plt.imshow(imgGray), plt.title('Gray Image')
plt.subplot(143), plt.imshow(imgBi), plt.title('Binary Image')
plt.subplot(144), plt.imshow(imgGrayBi), plt.title('Gray of Binary Image')
```

```python
contours, hierarchy = cv2.findContours(imgGrayBi, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# Sp xp cc contour theo din tch gim dn:
area_cnt = [cv2.contourArea(cnt) for cnt in contours]
area_sort = np.argsort(area_cnt)[::-1]
# Top 20 contour c din tch ln nht
area_sort[:20]
```

```python
def _drawBoundingBox(img, cnt):
  x,y,w,h = cv2.boundingRect(cnt)
  img = cv2.rectangle(img, (x,y),(x+w,y+h),(0,255,0),2)
  return img

imgOrigin = img.copy()
# V bounding box cho 25 contours c din tch ln nht
for i in area_sort[:25]:
  cnt = contours[i]
  imgOrigin = _drawBoundingBox(imgOrigin, cnt)
  
plt.imshow(imgOrigin)
```

```python
import numpy as np

def non_max_suppression(boxes, overlapThresh):
  '''
  boxes: List cc bounding box
  overlapThresh: Ngng overlapping gia cc hnh nh
  '''
  # Nu khng c bounding boxes th tr v empty list
  if len(boxes)==0:
    return []
  # Nu bounding boxes nguyn th chuyn sang float.
  if boxes.dtype.kind == "i":
    boxes = boxes.astype("float")

  # Khi to list ca index c la chn
  pick = []

  # Ly ra ta  ca cc bounding boxes
  x1 = boxes[:,0]
  y1 = boxes[:,1]
  x2 = boxes[:,2]
  y2 = boxes[:,3]

  # Tnh ton din tch ca cc bounding boxes v sp xp chng theo th t t bottom-right, chnh l ta  theo y ca bounding box
  area = (x2 - x1 + 1) * (y2 - y1 + 1)
  idxs = np.argsort(y2)
  # Khi to mt vng while loop qua cc index xut hin trong indexes
  while len(idxs) > 0:
    # Ly ra index cui cng ca list cc indexes v thm gi tr index vo danh sch cc indexes c la chn
    last = len(idxs) - 1
    i = idxs[last]
    pick.append(i)

    # Tm cp ta  ln nht (x, y) l im bt u ca bounding box v ta  nh nht (x, y) l im kt thc ca bounding box
    xx1 = np.maximum(x1[i], x1[idxs[:last]])
    yy1 = np.maximum(y1[i], y1[idxs[:last]])
    xx2 = np.minimum(x2[i], x2[idxs[:last]])
    yy2 = np.minimum(y2[i], y2[idxs[:last]])

    # Tnh ton width v height ca bounding box
    w = np.maximum(0, xx2 - xx1 + 1)
    h = np.maximum(0, yy2 - yy1 + 1)

    # Tnh ton t l din tch overlap
    overlap = (w * h) / area[idxs[:last]]

    # Xa index cui cng v index ca bounding box m t l din tch overlap > overlapThreshold
    idxs = np.delete(idxs, np.concatenate(([last],
      np.where(overlap > overlapThresh)[0])))
  # Tr ra list cc index c la chn
  return boxes[pick].astype("int")
 
boundingBoxes = [cv2.boundingRect(cnt) for cnt in contours]
boundingBoxes = np.array([(x,y,x+w,y+h) for (x,y,w,h) in boundingBoxes])
pick = non_max_suppression(boundingBoxes, 0.5)
```

```python
imgOrigin = img.copy()
for (startX, startY, endX, endY) in pick:
    imgOrigin = cv2.rectangle(imgOrigin, (startX, startY), (endX, endY), (0, 255, 0), 2)
plt.imshow(imgOrigin)
```
