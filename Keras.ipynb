```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical

#Data loader
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#Build model

model= Sequential()
model.add(Dense(700, activation='relu', input_shape=(28*28,)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
#Ham .add(): Them cac layer vao model


#Compile cho model
'''
    - Loss function:
    - OPtimizer:
    - 
'''
#xac dinh loss function, optimizer va metrics
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])


#Chuan hoa du lieu:
'''
Trước khi train, tôi cần reshape dữ liệu thành kích thước mà mạng đã được cấu hình (784,) 
và chia tỷ lệ của các giá trị nằm trong khoảng [0,1]. Các hình ảnh ở tập train được lưu 
trữ trong 1 mảng với shape là (60000,28,28) kiểu uint8 với các giá trị trong khoảng [0,255]. 
Tôi sẽ chuyển nó thành một mảng float32 với shape (60000,784) với các giá trị nằm trong khoảng [0,1]
'''
#reshape va chia ti le cac gia tri:
x_train = x_train.reshape(-1,784)
x_test = x_test.reshape(-1,784)
y_train = y_train.astype('float32') / 255.0
y_test = y_test.astype('float32') / 255.0

#Chuyen cac nhan sang so:
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)


#Huan Luyen va danh gia model:
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```
